```sos
// Imports 
var conv1d = require('./javascript/convolution');
var Plot = require('plotly-notebook-js');
```

```sos
// Define Moving Average filter Function
function hMovingAverage (taps)
{
    return Array(taps).fill(1).map(x => x/taps);
    
}
```

```sos
var h = hMovingAverage(11);
```

```sos
// Signals
var x = Array.apply(null, {length: 100}).map(Number.call, Number);
var x_noise = x.map(a => a + 2*((Math.random() * 2)-1));
var filtered = conv1d.Convolution1D(x_noise,h, true);
```

```sos
// Plot
var trace1 = {
  type: 'scatter',
  x: x,
  y: x_noise,
  name: 'Ramp + Noise',
  mode: 'markers',
  
};

var trace2 = {
  type: 'scatter',
  x: x,
  y: x,
  name: 'Ramp',
  mode: 'lines',
  
};

var trace3 = {
  type: 'scatter',
  x: x,
  y: filtered,
  name: 'Filtered Ramp+Noise',
  mode: 'lines',
  
};


var data = [ trace1 , trace2, trace3];

var layout = {
  title:'Moving Average Filter',
  xaxis: {
    title: 'x'
  },
  yaxis:{
      title: 'y'
  },
  margin: {
    l: 40,
    r: 30,
    b: 40,
    t: 40
  },
  legend: {
    font: {
      size: 10
    }
  },
  width: 900,
  height: 900,
  paper_bgcolor: 'rgb(254, 247, 234)',
  plot_bgcolor: 'rgb(254, 247, 234)',
  hovermode: 'closest',
  showlegend: true
};

var myPlot = Plot.createPlot(data, layout);


$$html$$ = myPlot.render();
```
