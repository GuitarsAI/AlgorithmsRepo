## Moving Average Filter Function

```python
# Moving Average Impulse Response
def hMovingAverage(taps):
    return (1/taps)*np.ones(taps)
```

```python
h=hMovingAverage(11)
```

## Convolution 1D with Delay Compensation

```python
# Define Convolution1D
def Convolution1D(f,g, delay=True):
    y=np.zeros(f.size+g.size-1)
    for n in range(f.size):
        for m in range(g.size):
                y[n+m]+=f[n]*g[m]
    if delay:
        return y[round((g.size-1)/2):-(g.size-1)]
    else:
        return y
```

## Define Signals

```python
x=np.linspace(0,100,100)
noise=np.random.uniform(-1,1,100)
x_noise = x+4*noise
x_noise_filtered=Convolution1D(x_noise,h)
```

## Plot

```python
plt.figure(figsize=(8,6))
plt.title("Moving Average Filter")
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, label='Ramp')
plt.plot(x_noise, 'g.', label='Ramp+Noise')
plt.plot(x_noise_filtered, 'r--', label='Filtered Ramp+Noise (Delay Compensated)')
plt.legend();
```
